# 5. Combining & Reshaping Data

```python
# make the second dataset

department_info = pd.DataFrame({
    "Department": ["HR", "IT", "Finance", "Marketing"],
    "Manager": ["Sarah", "Michael", "James", "Emma"]
})

print(department_info)
```

```python
# Inner Join (Default)
merged = pd.merge(df, department_info, on="Department")

print(merged)
```

```python
# Left Join
pd.merge(df, department_info,
         on="Department",
         how="left")
```

```python
# Right Join
pd.merge(df,
         department_info,
         on="Department",
         how="right")
```

```python
# Outer Join
pd.merge(df,
         department_info,
         on="Department",
         how="outer")
```

```python
# combines two or more DataFrames or Series by stacking them vertically (rows) or horizontally (columns).

pd.concat([df, department_info])
```

```python
# sets one or more existing columns as the index of a DataFrame.

df.set_index("Name")
```

```python
# resets the index of a DataFrame to the default integer index and optionally converts the current index into a column.

df.reset_index()
```

# 6. Mathematical & Element-wise Operations

```python
# performs element-wise addition between DataFrames or Series, automatically aligning data by index and column labels.

df["Salary"].add(1000)
```

```python
df["Age"].add(df["Salary"])
```

```python
# performs element-wise subtraction between DataFrames or Series, automatically aligning data by index and column labels.

df["Salary"].sub(df["Age"])
```

```python
# performs element-wise multiplication between DataFrames or Series, automatically aligning data by index and column labels.

df["Age"].mul(df["Salary"], fill_value=1)
```

```python
# performs element-wise division between DataFrames or Series, automatically aligning data by index and column labels.

df[["Age", "Salary"]].div(10)
```
