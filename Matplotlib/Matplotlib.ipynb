```python
# Create a circle that should look like a circle
theta = np.linspace(0, 2*np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Default aspect ratio (stretched)
ax1.plot(x, y, 'b-', linewidth=2)
ax1.set_title('Default Aspect (Stretched Circle)')
ax1.grid(True)

# Plot 2: Equal aspect ratio (true circle)
ax2.plot(x, y, 'r-', linewidth=2)
ax2.set_aspect('equal')
ax2.set_title('Equal Aspect (True Circle)')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Weather data example: Temperature (Celsius) and Rainfall (mm)
days = np.arange(1, 31)
temperature = 20 + 5 * np.sin(days/5) + np.random.randn(30) * 2
rainfall = np.random.poisson(lam=5, size=30)

fig, ax1 = plt.subplots(figsize=(10, 6))

# Temperature on left y-axis
color1 = 'tab:red'
ax1.set_xlabel('Day of Month')
ax1.set_ylabel('Temperature (°C)', color=color1)
ax1.plot(days, temperature, color=color1, marker='o', linewidth=2)
ax1.tick_params(axis='y', labelcolor=color1)

# Create second y-axis for rainfall
ax2 = ax1.twinx()
color2 = 'tab:blue'
ax2.set_ylabel('Rainfall (mm)', color=color2)
ax2.bar(days, rainfall, color=color2, alpha=0.3)
ax2.tick_params(axis='y', labelcolor=color2)

ax1.set_title('Daily Temperature and Rainfall')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a complex figure with many subplots
fig, axes = plt.subplots(3, 3, figsize=(10, 8))

# Add random data to each subplot
for i, ax in enumerate(axes.flatten()):
    ax.plot(np.random.randn(20).cumsum())
    ax.set_title(f'Plot {i + 1}')
    ax.set_xlabel('Time')
    ax.set_ylabel('Value')

# Adjust subplot spacing manually
plt.subplots_adjust(
    hspace=0.6,
    wspace=0.4
)

print("Subplot spacing adjusted successfully.")

plt.show()
```
